## 2. Implementaion

Your task is to implement near duplicate detection using LSH with cosine similarity.
More specifically you have to:
* Generate duplicate **candidates** based on LSH with $b$ bands and $r$ rows per band
* Refine the candidates by computing the exact cosine distance
* Report all pairs/duplicates with cosine distance < $d$

Implement a function that computes the cosine distance between two rows (instances) in the data.

```python
def cosine_distance(X, i, j):
    """Compute cosine distance between two rows of a data matrix.
    
    Parameters
    ----------
    X : np.array, shape [N, D]
        Data matrix.
    i : int
        Index of the first row.
    j : int
        Index of the second row.
        
    Returns
    -------
    d : float
        Cosine distance between the two rows of the data matrix.
        
    NOTE: this function is actually not used due to the way we wrote the other functions
    (IE stored referenced to actual data not indexes in initial matrix) 
    This function is provided just to complete all the tasks. 
    See cosine_distance_alternative function for the alternative used in the code.
    """
    d = 1.0 - np.dot(X[i,:],X[j,:])/(np.linalg.norm(X[i,:])*np.linalg.norm(X[j,:]))
    return d

def cosine_distance_alternative(v1,v2):
    """
    TODO: documentation
    """
    d = np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
    return d
               
```

```python
def batchGetCandidates(batchSize,X):
    """
    Returns all Candidate pairs found during running a batch 
    """
    dictCandidates = {}
    candidatePairs = set()
    
    randVects = np.random.rand(batchSize,X.shape[1])-0.5 #IE uniform distribution from -0.5 to 0.5
    for i in range(X.shape[0]):
        sig = batchGetSongSignature (X,i,randVects)
        sig_string = str(sig) #as strings are hashable by default
        if sig_string in dictCandidates:
            dictCandidates[sig_string ].append(i)
        else:
            dictCandidates.update( {sig_string  : [i]} )
    
    for key, bucket in dictCandidates.items():
        if len(bucket) > 1:
            candidatePairs.update(tuple(x for x in it.combinations(bucket,2)))
    
    #print("debug. CandidatePairs is : " + str (candidatePairs))
    return candidatePairs
    

def batchGetSongSignature(X ,i ,rand_vects):
    """
    returns the data's signature(IE a column of the sketch) based on the data (np.array)
    and a matrix containing random vectors (number of vectors depends on size of batch)
    to be used with random hyperplane hashing technique
    """
    return [hyperplaneHash(X, i,rand_vect) for rand_vect in rand_vects]

def hyperplaneHash(X, i, v2):
    """
        returns +1 or -1, depending on v1*v2. v1 is line i of X
        The basis of implementing LSH for cosine distance
    """
    if (np.dot(X[i,:],v2) >= 0.0):
        return 1
    else:
        return -1
    
def refineStep(candidates,d,X):
    duplicates = set ()
    for pair in candidates:
        #print("debug. pair is: " + str(pair))
        dist = cosine_distance(X,pair[0],pair[1])
        if dist <= d :
            duplicates.add(pair + (dist,))
    return duplicates


def LSH(X, b=8, r=32, d=0.3):
    """Find candidate duplicate pairs using LSH and refine using exact cosine distance.
    
    Parameters
    ----------
    X : np.array shape [N, D]
        Data matrix.
    b : int
        Number of bands.
    r : int
        Number of rows per band.
    d : float
        Distance treshold for reporting duplicates.
    
    Returns
    -------
    duplicates : {(ID1, ID2, d_{12}), ..., (IDX, IDY, d_{xy})}
        A set of tuples indicating the detected duplicates.
        Each tuple should have 3 elements:
            * ID of the first song
            * ID of the second song
            * The cosine distance between them
    
    n_candidates : int
        Number of detected candidate pairs.
        
    """
    np.random.seed(158)
    n_candidates = 0
    candidates = set()
    
    for _ in range(b):
        batchCandidates = batchGetCandidates(r,X)
        #now check if the batchCandidates are already in candidates. if not, add them:
        candidates.update(batchCandidates) #luckiliy python sets are smart
    
    n_candidates = len(candidates)
    duplicates = refineStep(candidates,d,X)
                               
    return duplicates, n_candidates
```

## 3. Compare runtime

Your task is to implement code for runtime comparison between LSH and the naive nested for loop implementation.

```python
# naively compute the duplicates using a double for loop
def naive_duplicates(X, d = 0.3):
    """
    Parameters
    ----------
    X : np.array, shape [N, D]
        Data matrix.
    d : float
        Distance treshold for reporting duplicates.
    
    Returns
    -------
    duplicates : {(ID1, ID2, d_{12}), ..., (IDX, IDY, d_{xy})}
        A set of tuples indicating the detected duplicates.
        Each tuple should have 3 elements:
            * ID of the first song
            * ID of the second song
            * The cosine distance between them
    """
    N = X.shape[0]
    duplicates = set()
    for i in range(N):
        for j in range(N):
            d_ij = cosine_distance(X, i, j)
            if d_ij < d and i != j:
                duplicates.add((i, j, d_ij))
    return duplicates
```

```python
def runtime_comparison():
    """
    Compare the runtime between LSH and the naive approach.
    
    Returns
    -------
    trace : [(n1, lsh_dur, naive_dur), (n2, lsh_dur, naive_dur), ... ]
            A list of tuples with execution times for different number of songs.
            Each tuple should have 3 elements:
                * number of songs considered
                * duration of the LSH approach
                * duration of the naive approach
    s"""
    trace = []
    for n in np.arange(25, 351, 25):
        print('Running comparison for {} songs.'.format(n))
        
        ### YOUR CODE HERE ###
        start = time.time()
        dupli_lsh,_ = LSH(scaled_data[0:n,:])
        end = time.time()
        lsh_dur = end-start
        start = time.time()
        dupli_naive = naive_duplicates(scaled_data[0:n,:])
        end = time.time()
        naive_dur = end-start
        
        if len(dupli_lsh) != len(dupli_naive):
            print("Number of missed pairs! " + str(len(dupli_naive)-len(dupli_lsh)))
        
        trace.append((n,lsh_dur,naive_dur))
        
        
    return trace
```

```python
trace = runtime_comparison()
```

Plot the differecene in runtime. On the x-axis plot the number of songs processed and on the y-axis plot the runtime in seconds for both approaches. You should obtain a plot similar to the one shown below.

```python
trace
x =[]
y1=[]
y2=[]
for measure in trace:
    x.append(measure[0])
    y1.append(measure[1])
    y2.append(measure[2])

plt.figure()
plt.plot(x,y1,label="LSH")
plt.plot(x,y2,label="naive")

plt.title("Runtime Comparison")
plt.xlabel("Number of songs (number)")
plt.ylabel("Time taken (seconds)")
plt.legend(loc='upper left')    


plt.show
```
